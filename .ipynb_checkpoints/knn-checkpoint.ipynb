# Classifier Documentation

This classifier demonstrates a simple example dataset creation and is typically used as a starting point for implementing classification algorithms such as k-Nearest Neighbors (k-NN).

## Dataset Description

The function `createDataSet()` generates a small sample dataset consisting of four points in a 2-dimensional feature space. Each point is assigned a label indicating its class membership.

- The dataset `group` is a NumPy array of shape `(4, 2)` containing four points.
- The corresponding `labels` list contains the class labels (`'A'` or `'B'`) for each point.

This dataset can be used to test classification algorithms quickly.

## Euclidean Distance

A fundamental concept used in many classifiers, especially distance-based classifiers like k-NN, is the **Euclidean distance** between data points.

The Euclidean distance $$ d $$ between two points $$ \mathbf{x} = (x_1, x_2, \ldots, x_n) $$ and $$ \mathbf{y} = (y_1, y_2, \ldots, y_n) $$ in an $$ n $$-dimensional space is defined as:

$$
d(\mathbf{x}, \mathbf{y}) = \sqrt{\sum_{i=1}^{n} (x_i - y_i)^2}
$$

where:
- $$ x_i $$ and $$ y_i $$ are the coordinates of the two points in the $$ i^{th} $$ dimension,
- $$ n $$ is the number of dimensions (features).

This distance metric measures the straight-line (or "as-the-crow-flies") distance between points in the feature space.

````python
import matplotlib.pyplot as plt
from numpy import array

# Get dataset
group, labels = createDataSet()

# Convert labels to colors for plotting
colors = ['red' if label == 'A' else 'blue' for label in labels]

# Prepare the scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(group[:, 0], group[:, 1], c=colors, s=100, marker='o', edgecolor='k')

for i, label in enumerate(labels):
    plt.text(group[i, 0] + 0.02, group[i, 1], label, fontsize=14)

plt.title('Sample Classifier Data Points')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.axis('equal')
plt.show()
````
