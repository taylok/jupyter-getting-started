```python
# factorialA: iterative using a for loop
def factorialA(n):
    result = 1
    for i in range(2, n+1):
        result *= i
    return result

# factorialB: recursive implementation
def factorialB(n):
    if n <= 1:
        return 1
    else:
        return n * factorialB(n-1)
```

```python
%timeit factorialA(100)
```

```python
%timeit factorialB(100)
```
